# Exercício 01
<p style="text-align:justify;">
Crie um programa que armazene um vetor com as notas de dez alunos, calcule e imprima a média dessas notas e depois informe a quantidade de notas acima e abaixo da média calculada.
</p>

```java
import java.util.Scanner;

public class MediaNotas {
    public static void main(String[] args) {
        // Criando um vetor para armazenar as notas.
        double[] notas = new double[10];
        
        // Criando um Scanner para receber as notas.
        Scanner scanner = new Scanner(System.in);
        
        // Pedindo ao usuário para inserir as notas
        System.out.println("Digite as notas dos dez alunos: ");
        for (int i = 0; i < 10; i++) {
            System.out.print("Nota do aluno " + (i + 1) + ": ");
            notas[i] = scanner.nextDouble();
        }
        
        // Calculando a média das notas.
        double soma = 0;
        for (int i = 0; i < 10; i++) {
            soma += notas[i];
        }
        double media = soma / 10;
        
        // Imprimindo a média das notas
        System.out.println("A média das notas é: " + media);
        
        // Contando as notas acima e abaixo da média.
        int acimaMedia = 0;
        int abaixoMedia = 0;
        for (int i = 0; i < 10; i++) {
            if (notas[i] > media) {
                acimaMedia++;
            } else if (notas[i] < media) {
                abaixoMedia++;
            }
        }
        
        // Imprimindo a quantidade de notas acima e abaixo da média.
        System.out.println("Quantidade de notas acima da média: " + acimaMedia);
        System.out.println("Quantidade de notas abaixo da média: " + abaixoMedia);
        
        // Fechando o scanner
        scanner.close();
    }
}
```

# Exercício 02
<p style="text-align:justify;">
Construa uma classe em Java que seja capaz de efetuar a leitura de uma matriz de valores reais de tamanho 3 x 3, calculando seu determinante.
</p>

```java
import java.util.Scanner;
public class CalculadoraDeterminante02 {
    public static void main(String[] args) {
        Scanner scanner = new Scanner(System.in);

        // Ler a matriz 3x3.
        double[][] matriz = new double[3][3];
        System.out.println("Digite os elementos da matriz 3x3: ");
        for (int i = 0; i < 3; i++) {
            for (int j = 0; j < 3; j++) {
                matriz[i][j] = scanner.nextDouble();
            }
        }

        // Calcular o determinante.
        double determinante = calcularDeterminante(matriz);
        System.out.println("O determinante da matriz é: " + determinante);

        scanner.close();
    }

    // Método para calcular o determinante de uma matriz 3x3
    public static double calcularDeterminante(double[][] matriz) {
        double determinante = 0;
        determinante += matriz[0][0] * (matriz[1][1] * matriz[2][2] - matriz[1][2] * matriz[2][1]);
        determinante -= matriz[0][1] * (matriz[1][0] * matriz[2][2] - matriz[1][2] * matriz[2][0]);
        determinante += matriz[0][2] * (matriz[1][0] * matriz[2][1] - matriz[1][1] * matriz[2][0]);
        return determinante;
    }
}
```

# Exercício 03
<p style="text-align:justify;">
Faça uma classe contendo dois arrays do tipo inteiro, um chamado par e outro chamado ímpar. Cada array deve conter 10 elementos. A seguir, faça um laço de repetição com 10 iterações e que contenha internamente a geração randômica de números entre 1 e 20. Se o valor gerado for par, armazene no array chamado par, caso contrário, no array chamado impar. Ao final, apresente o conteúdo de cada um dos arrays.
</p>

```java
import java.util.Random;
public class NumerosParesImpares03 {
    public static void main(String[] args) {
        int[] par = new int[10];
        int[] impar = new int[10];
        Random random = new Random();

        for (int i = 0; i < 10; i++) {
            int numero = random.nextInt(20) + 1; // Gera um número aleatório entre 1 e 20.

            if (numero % 2 == 0) {
                par[i] = numero;
            } else {
                impar[i] = numero;
            }
        }

        // Imprimindo os números pares.
        System.out.println("Números Pares: ");
        for (int num : par) {
            if (num != 0) {
                System.out.print(num + " ");
            }
        }

        // Imprimindo os números ímpares.
        System.out.println("\nNúmeros Ímpares: ");
        for (int num : impar) {
            if (num != 0) {
                System.out.print(num + " " );
            }
        }
    }
}
```

# Exercício 04
<p style="text-align:justify;">
Usando um array unidimensional de 10 elementos do tipo String, faça uma classe que simule a ocupação de um estacionamento de veículos. Considere que a posição do array corresponde ao número da vaga e para cada vaga é armazenada a placa do veículo que está estacionado. Sua classe deve conter um laço de repetição simulando um menu que controle a entrada e a saída dos veículos com as opções: 1-Entrada, 2-Saída e 3-Listar situação atual e 4-Encerrar o programa. Se o usuário escolher 1, solicite o número da vaga e a placa do veículo; se escolher 2, solicite o número da vaga que será liberada e armazene o valor branco na posição correspondente ao array; se escolher 3, liste a situação atual apresentando em tela todos os elementos do array; quando for 4, encerre o laço e o programa.
</p>

```java
import java.util.Scanner;

public class EstacionamentoSimples04 {

    private static final int NUMERO_VAGAS = 10;
    private static final String[] vagas = new String[NUMERO_VAGAS];

    private static Scanner scanner = new Scanner(System.in);

    public static void main(String[] args) {
        boolean executando = true;
        while (executando) {
            exibirMenu();
            int opcao = scanner.nextInt();
            executarOpcao(opcao);
        }
        scanner.close();
    }

    private static void exibirMenu() {
        System.out.println("\nMenu:");
        System.out.println("1- Entrada");
        System.out.println("2- Saída");
        System.out.println("3- Listar situação atual");
        System.out.println("4- Encerrar o programa");
        System.out.print("Escolha uma opção: ");
    }

    private static void executarOpcao(int opcao) {
        switch (opcao) {
            case 1:
                entrada();
                break;
            case 2:
                saida();
                break;
            case 3:
                listarSituacaoAtual();
                break;
            case 4:
                encerrarPrograma();
                break;
            default:
                System.out.println("Opção inválida! Por favor, escolha uma opção válida.");
                break;
        }
    }

    private static void entrada() {
        System.out.print("Digite o número da vaga: ");
        int numeroVagaEntrada = scanner.nextInt();
        scanner.nextLine(); // Consumir a quebra de linha.
        if (numeroVagaEntrada >= 0 && numeroVagaEntrada < NUMERO_VAGAS) {
            System.out.print("Digite a placa do veículo: ");
            String placa = scanner.nextLine();
            vagas[numeroVagaEntrada] = placa;
            System.out.println("Veículo estacionado com sucesso na vaga " + numeroVagaEntrada);
        } else {
            System.out.println("Vaga inválida!");
        }
    }

    private static void saida() {
        System.out.print("Digite o número da vaga que será liberada: ");
        int numeroVagaSaida = scanner.nextInt();
        if (numeroVagaSaida >= 0 && numeroVagaSaida < NUMERO_VAGAS) {
            vagas[numeroVagaSaida] = null;
            System.out.println("Veículo liberado da vaga " + numeroVagaSaida);
        } else {
            System.out.println("Vaga inválida!");
        }
    }

    private static void listarSituacaoAtual() {
        System.out.println("Situação atual do estacionamento:");
        for (int i = 0; i < NUMERO_VAGAS; i++) {
            if (vagas[i] != null) {
                System.out.println("Vaga " + i + ": " + vagas[i]);
            } else {
                System.out.println("Vaga " + i + ": Vaga livre");
            }
        }
    }

    private static void encerrarPrograma() {
        System.out.println("Encerrando o programa...");
        scanner.close();
        System.exit(0);
    }
}
```

# Exercício 05
<p style="text-align:justify;">
Uma imagem é formada por pixels. Considere uma imagem com dimensão de 40 x 40 e faça uma classe que contenha um array bidimensional com essas dimensões. A seguir, para cada posição
desse array bidimensional armazene um valor aleatório entre 0 e 255 (esses valores correspondem às tonalidades aplicadas sobre a imagem). Apresente em tela os 1600 valores gerados.
</p>

```java
import java.util.Random;

public class ImagemPixels {
    public static void main(String[] args) {
        int[][] imagem = new int[40][40];
        Random random = new Random();

        // Preenchendo a imagem com valores aleatórios entre 0 e 255.
        for (int i = 0; i < 40; i++) {
            for (int j = 0; j < 40; j++) {
                imagem[i][j] = random.nextInt(256); // Gera um número aleatório entre 0 e 255.
            }
        }

        // Apresentando os valores gerados em tela
        System.out.println("Valores dos pixels da imagem: ");
        for (int i = 0; i < 40; i++) {
            for (int j = 0; j < 40; j++) {
                System.out.print(imagem[i][j] + " ");
            }
            System.out.println();
        }
    }
}
```
